### Feature Engineering

```python
# seperating object and numerical features 
obj=[]
flt=[]
for i in columns:
    if data[i].dtype == 'O':
        obj.append(i)
    else:
        flt.append(i)
```

```python
# imputing all the null values in both Object and Numrical features
from sklearn.impute import SimpleImputer
si=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
data[obj] = si.fit_transform(data[obj])
si=SimpleImputer(missing_values=np.nan,strategy='median')
data[flt] = si.fit_transform(data[flt])
```

here, we can see sex and Thyroid having ['F','M'] and ['P', 'N'] unique values,respectively. otherwise we will be having ['f', 't'] as unique values

```python
# replacing those values from sex and tyroid
data['sex'].replace({'F':0,'M':1},inplace=True)
data['Thyroid'].replace({'N':0,'P':1},inplace=True)
```

```python
# again seperating features in obj and numerical 
obj=[]
flt=[]
for i in columns:
    if data[i].dtype == 'O':
        obj.append(i)
    else:
        flt.append(i)
```

```python
# replacing others
data[obj]=data[obj].replace({'f':0,'t':1})
```

### Model trainning

```python
# seperating independent and dependent variables
x = data.drop('Thyroid',axis = 1)
y = data.Thyroid
```

```python
# applying train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=40,test_size=.30)
```

```python
# displaying shapes
print('x train shape : ',x_train.shape)
print('y train shape : ',y_train.shape)
print('x test shape : ',x_test.shape)
print('y test shape : ',y_test.shape)
```

```python
cl=x_train.columns
cl
```

#### Transformations

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test= sc.transform(x_test)
```

```python
# importing model
from sklearn.ensemble import ExtraTreesClassifier
etc=ExtraTreesClassifier(n_estimators=1000,criterion='gini')
```

```python
# fitting model
etc.fit(x_train,y_train)
```

```python
# predicting output
y_pred=etc.predict(x_test)
```

### Scoring

```python
# importing and displaying classification report
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
# checking best features
feature_imp=pd.Series(etc.feature_importances_,index=cl)
feature_imp*100
```

```python
flt
```

#### Kfold scoring

```python
from sklearn.model_selection import StratifiedKFold,cross_val_score
skf=StratifiedKFold(n_splits=10,random_state=40,shuffle=True)

score=cross_val_score(estimator=etc,X=x_train,y=y_train,cv=skf)
```

```python
print('this is an actual score for our model: ',score.mean()*100)
```
